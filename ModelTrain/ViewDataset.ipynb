```python
from CaptchaDataset import Dataset
import matplotlib.pyplot as plt
```

```python
ds = Dataset()
train_name, train_labels = ds.load_labels('datasets/train/labels.csv')
val_name, val_labels = ds.load_labels('datasets/val/labels.csv')

train_gen = ds.image_generator('datasets/train', name=train_name, labels=train_labels)
val_labels = ds.image_generator('datasets/val', name=val_name, labels=val_labels)
```

```python
for image in train_gen:
    # (data, label)
    img_arr, label_arr = image
    break
```

```python
def captcha_show(images_arr, labels_arr):
    fig, axes = plt.subplots(6, 1, figsize=(5, 15))
    axes = axes.flatten()
    for img, ax, label in zip(images_arr, axes, labels_arr):
        ax.imshow(img)
        # label是字节串格式，需要解码为utf-8
        ax.set_title(f'label: {label.numpy().decode("utf-8")}')

        ax.set_xticks([])
        ax.set_yticks([])

    plt.show()
captcha_show(img_arr, label_arr)
```
