```python
import tensorflow as tf
from tensorflow import keras
from CaptchaDataset import Dataset
```

```python
IMG_HEIGHT = 60
IMG_WIDTH = 160
BATCH_SIZE = 32

ds = Dataset()
```

```python
test_name, test_label = ds.load_labels('./datasets/test/labels.csv')
AUTOTUNE = tf.data.AUTOTUNE

test_dataset = ds.captcha_image_generator('datasets/test', test_name, test_label)
test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(AUTOTUNE)
```

```python
model:keras.Model = keras.models.load_model('./model/CNN_RNN_0073.keras')
model.trainable = False
```

```python
predictions = model.predict(test_dataset)
```

```python
def calculate_accuracy(predictions, test_label):
    correct = 0
    total = len(test_label)
    for i, pred in enumerate(predictions):
        # 将预测结果转为字符",
        pred_chars = ''
        for j in range(4):  # 对每个字符进行处理",
            char_idx = tf.argmax(pred[j]).numpy()
            if char_idx < 10:   # 数字",
                pred_chars += str(char_idx)
            elif char_idx < 36:
                pred_chars += chr(ord('a') + char_idx - 10)
            else:
                pred_chars += chr(ord('A') + char_idx - 36)
        # print(f'True: {test_label[i]}, Predict: {pred_chars}')
        if pred_chars == test_label[i]:
            correct += 1
    return correct / total
```

```python
acc = calculate_accuracy(predictions, test_label)
print(f"测试集准确率：{acc:.2%}")
```
